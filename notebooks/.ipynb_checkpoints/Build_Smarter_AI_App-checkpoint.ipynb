## Estructura de la respuesta de los LLM

La respuesta no siempre es un simple texto. En muchos casos (especialmente con Chat Models como DeepSeek), obtienes un objeto estructurado con mucha más información que pueden ser extraidas como objetos generados por el modelo que estes usando. 

EL objeto `content`, es el que tiene la respuesta del prompt, para que se muestre solo ese usamos `response.content`. Este viene formateado en Markdown. Para vizualizar correctamente el formato importamos librerias de vizualizacion de jupyter: 

````python
from IPython.display import display, Markdown
````

Definimos una funcion para reemplazar a `print()` en la respuesta que muestre el texto en el formato correcto: 

````python
def show(response):
    content = response.content if hasattr(response, "content") else response
    display(Markdown(content))
````

Ahora solamante llamamos a show(response) y devolvera `content` vizualizado correctamnente. 

### Recapitulando: 

En el Lab anterior se abordo los diferentes tipos de prompt, el metodo LCEL para crear `chains` y `templates` que generen interacciones reurtilizables con la IA y varios agentes de ejemplo en el siguiente ejemplo vemos un agente que aborda todos los temas tratados: 

````python
content = """
        El marketing digital es el conjunto de estrategias y técnicas que se utilizan para promocionar productos o servicios a través de canales digitales. Incluye herramientas como redes sociales, email marketing, SEO (optimización para motores de búsqueda) y publicidad online. El SEO se centra en mejorar la visibilidad de una web en los resultados orgánicos de buscadores como Google, mientras que la publicidad online permite llegar a audiencias específicas mediante anuncios pagados.
"""

question = "¿Qué técnica del marketing digital se centra en mejorar la visibilidad en buscadores?"

# Template
template = """
Responde a la siguiente {question} basándote en {content} proporcionado:
Si no estás seguro de la respuesta, responde: "No estoy seguro de la respuesta".

Respuesta:
"""

# Promt
prompt = PromptTemplate.from_template(template)

# llm
llm = get_llm(llm="dsk", params={"temperature": 0.5})

# parser
parser = StrOutputParser()

# Chain
qa_chain = (
    prompt
    | llm
    | parser
)

response = qa_chain.invoke({"question": question, "content": content})
show(response)
````

## Modelos de chat

Los **chat models** son una evolución práctica de los modelos de lenguaje tradicionales (LLM). En lugar de trabajar con una única cadena de texto (input → output), están diseñados para operar sobre **conversaciones estructuradas**, donde cada mensaje tiene un **rol definido**. Esto cambia completamente cómo se diseñan aplicaciones de IA.

Por debajo sigue siendo un LLM, pero con una capa de abstracción orientada a diálogo. En vez de enviarle un prompt plano, le envías una **lista de mensajes** con contexto acumulado.

Estructura típica:

* **system** → define comportamiento (reglas, tono, objetivos)
* **user** → input del usuario
* **assistant** → respuestas del modelo

Esto permite algo clave: **control explícito del contexto** ya que, el comportamiento del modelo es **más predecible y modular**.

Los chat models permiten construir sistemas más avanzados porque:

* Mantienen **memoria conversacional**
* Separan claramente **instrucciones vs. input**
* Facilitan el uso de **herramientas externas (tools)**
* Encajan de forma natural con frameworks como LangChain

### Roles

````python
[
  {"role": "system", "content": "..."},         # "system": Comportamiento del modelo
  {"role": "user", "content": "..."},           # "human" : Mensaje del usuario
  {"role": "assistant", "content": "..."},      # "ai": Respuesta del modelo
  {"role": "assistant", "tool_calls": [...]},   # ToolMessage: Llamada a herramienta
  {"role": "tool", "content": "..."}            # ToolMessage: Respuesta de la herramienta
]
````

### Integracion con LCEL

````python
# 1. Prompt con variables
prompt = ChatPromptTemplate.from_messages([
    ("system", "Eres un experto en marketing data-driven. Da respuestas breves, prácticas y accionables en ."),
    ("human", "Contexto: {context}\n\nPregunta: {input}")
])

context = "Una tienda online tiene mucho tráfico pero baja tasa de conversión."
input = "¿Qué optimizaciones recomiendas?"

# llm
llm = get_llm(llm="dsk", params={"temperature": 0.5})

# Chain, no necesitamos parser porque los modelos de chat devuelven el mensaje estruturado. 
chain = prompt | llm

# Ejecución
response = chain.invoke({
    "context": context,
    "input": input
})

# Usamos la funcion de vizualizacion de Jupiter para formatear el markdown
show(response)
````

Otro ejemplo: 

````python
# llm
llm = get_llm(llm="dsk", params={"temperature": 0.5})

# Cadenas de mensajes
response = llm.invoke(
    [
        SystemMessage(content="Eres un consultor experto en marketing digital y analítica. Responde siempre con recomendaciones claras y accionables en una sola frase."),
        
        HumanMessage(content="Tengo mucho tráfico en mi web pero pocas conversiones, ¿qué podría estar pasando?"),
        
        AIMessage(content="Probablemente tienes fricción en el funnel o una propuesta de valor poco clara."),
        
        HumanMessage(content="¿Qué debería analizar primero?")
    ]
)

show(response)
````

## MessagesPlaceholder (contexto dinámico)

En LangChain, MessagesPlaceholder es un componente dentro de los ChatPromptTemplate que permite insertar dinámicamente una lista completa de mensajes en una posición específica del prompt. 

A diferencia de las variables simples como {input}, que solo sustituyen texto, MessagesPlaceholder actúa como un contenedor para múltiples mensajes estructurados (por ejemplo, HumanMessage, AIMessage o SystemMessage), lo que lo convierte en una herramienta fundamental para gestionar conversaciones complejas. 

Su principal utilidad es incorporar historial de diálogo, contexto dinámico o secuencias de interacción sin tener que reconstruir manualmente cada mensaje en el template. De este modo, facilita la creación de aplicaciones conversacionales más avanzadas, como chatbots con memoria, sistemas de preguntas y respuestas contextuales o flujos de interacción multi-turno, manteniendo una arquitectura limpia, modular y escalable.

Ejemplo: 

````python
# Modelo
llm = get_llm(llm="dsk", params={"temperature": 0.5})

# Prompt con placeholder
prompt = ChatPromptTemplate.from_messages([
    ("system", "Eres un experto en marketing digital. Responde de forma breve, clara y práctica."),
    MessagesPlaceholder("msgs")
])

# Input dinámico
input_ = {
    "msgs": [
        HumanMessage(content="Tengo muchas visitas pero pocas ventas, ¿qué hago?")
    ]
}

# Chain
chain = prompt | llm

# Ejecución
response = chain.invoke(input_)
show(response)
````

## Output Parsers en LangChain

En LangChain, un output parser es un componente que toma la salida del LLM (texto) y la transforma en un formato estructurado. Esto es clave porque los modelos, por defecto, devuelven texto libre (muchas veces en markdown), que no es directamente utilizable en código.

Los output parsers se utilizan cuando necesitas:

- estructurar datos (JSON, listas, tablas…)
- normalizar respuestas
- integrar LLMs con sistemas (APIs, bases de datos, BI…)

#### Tipos comunes de parsers

LangChain tiene varios, pero en este contexto destacan:

- JSON Parser
    - Devuelve datos estructurados
    - Compatible con esquemas definidos (Pydantic)
    - Muy fiable sin necesidad de function calling
- CSV Parser
    - Devuelve listas separadas por comas
    - Útil para datos tabulares simples

### JSON Output Parser

El **JSON Output Parser** permite transformar la salida de un modelo de lenguaje en un objeto JSON estructurado y predecible, en lugar de texto libre. Su funcionamiento se basa en definir previamente un esquema que especifica exactamente qué campos debe contener la respuesta y qué tipo de datos espera en cada uno. 

A partir de ese esquema, el parser genera instrucciones que se incorporan al prompt para guiar al modelo a producir una salida válida, y posteriormente valida y convierte esa respuesta en un formato directamente utilizable en código. Esto lo convierte en una herramienta esencial cuando se trabaja con automatización, integración de sistemas o análisis de datos, ya que permite pasar de respuestas ambiguas a estructuras fiables y listas para ser procesadas por aplicaciones, APIs o herramientas de business intelligence.

Importar las librerias necesarias

````python
from langchain_core.output_parsers import JsonOutputParser
from pydantic import BaseModel, Field
````

#### Clase BaseModel

`BaseModel` es una clase de la librería **Pydantic** que se utiliza para definir y validar estructuras de datos de forma clara y tipada en Python. Permite crear “modelos” donde especificas qué campos debe tener la información, qué tipo de datos espera cada uno (por ejemplo, texto, número, lista, etc.) y, opcionalmente, añadir descripciones o reglas. 

Cuando se usa en contextos como LangChain, `BaseModel` actúa como un esquema que garantiza que la salida del modelo siga una estructura concreta, facilitando así la validación automática y la conversión de datos en formatos fiables y utilizables en aplicaciones. En este caso: 
```python
{
  "setup": "...",
  "punchline": "..."
}
```

Al crear una clase que hereda de `BaseModel`, se establece un esquema claro con campos específicos (como `setup` y `punchline`) y sus tipos de datos, lo que permite validar automáticamente la respuesta generada. Además, el uso de `Field(description=...)` no solo documenta cada campo, sino que también guía al modelo sobre qué contenido debe incluir en cada uno, mejorando la calidad del resultado. 

Si se utiliza la forma simple 
```python
JsonOutputParser()
```
El parse funciona sin esquemas propios, menos fialble y estructura mas simple. Por ejemplo: 
```python
{
  "respuesta": "...",
  "detalle": "..."
}
```

Crear parse jJSON estructurado: 

````python
class Chiste(BaseModel):
    setup: str = Field(description="pregunta o planteamiento para iniciar el chiste")
    punchline: str = Field(description="respuesta o conclusión que resuelve el chiste")
````

Contrinua con el ejemplo: 

````python
# Consulta destinada a pedirle al modelo de lenguaje que rellene la estructura de datos
joke_query = "Tell me a joke."

# Configura un parser e inyecta las instrucciones en la plantilla del prompt
output_parser = JsonOutputParser(pydantic_object=Chiste)

# Obtiene las instrucciones de formato para el parser
# Esto genera texto guía que indica al LLM cómo debe formatear su respuesta
format_instructions = output_parser.get_format_instructions()

# Crea una plantilla de prompt que incluye:
# 1. Instrucciones para que el LLM responda a la consulta del usuario
# 2. Instrucciones de formato para asegurar que el LLM devuelva datos estructurados correctamente
# 3. El placeholder de la consulta del usuario
prompt = PromptTemplate(
    template="Responde a la consulta del usuario.\n{format_instructions}\n{query}\n",
    input_variables=["query"],  # Variables dinámicas que se proporcionarán al ejecutar la cadena
    partial_variables={"format_instructions": format_instructions},  # Variables estáticas definidas al crear el prompt
)

# llm
llm = get_llm(llm="dsk", params={"temperature": 0.5})

# Cadena
chain = prompt | llm | output_parser
chain.invoke({"query": joke_query})
````

### Comma-Separated List parser en LangChain,

El **Comma-Separated List Output Parser** en LangChain se utiliza cuando necesitas que el modelo genere una lista de elementos en un formato simple y fácilmente procesable, es decir, separados por comas. A diferencia del JSON parser, que está orientado a estructuras más complejas, este parser es ideal para casos donde solo necesitas colecciones de valores como categorías, nombres, etiquetas o recomendaciones. 

Su principal ventaja es la simplicidad: permite transformar directamente una respuesta textual en una lista de Python sin necesidad de definir esquemas avanzados, lo que lo convierte en una herramienta muy útil para tareas rápidas de clasificación, generación de ideas o extracción de elementos concretos a partir de una consulta.

Importar librerias necesarias

````python
from langchain_core.output_parsers import CommaSeparatedListOutputParser
````

Continua el ejemplo:

````python
# Crea una instancia del parser que convertirá texto separado por comas en una lista de Python
output_parser = CommaSeparatedListOutputParser()

# Estas instrucciones explican al LLM que debe devolver los elementos en formato separado por comas
format_instructions = output_parser.get_format_instructions()

# Prompt
prompt = PromptTemplate(
    template="Responde la consulta del usuario. {format_instructions}\n Lista 5 {subject}.",
    input_variables=["subject"],  # Esta variable se proporcionará cuando se ejecute la cadena
    partial_variables={"format_instructions": format_instructions},  # Esta variable se establece una sola vez al crear el prompt
)

llm = get_llm(llm="dsk", params={"temperature": 0.5})

# Cadena de procesamiento
chain = prompt | llm | output_parser

# Ejecutar
chain.invoke({"subject": "ice cream flavors"})
````

Este ejercicio se sitúa en el contexto de una empresa que quiere aprovechar la inteligencia artificial para analizar situaciones de negocio —por ejemplo, problemas de marketing, rendimiento de campañas o comportamiento de usuarios— de forma automática y consistente. 

La necesidad principal de la empresa no es obtener respuestas largas o explicativas, sino **información clara, estructurada y reutilizable**, que pueda integrarse fácilmente en sus sistemas, dashboards o procesos de toma de decisiones. Por ello, el objetivo del ejercicio es simular cómo un modelo de IA puede generar análisis estandarizados a partir de una descripción dada, de manera que cada respuesta siga siempre el mismo formato. 

* problema
* causa
* acción recomendada
* impacto

Esto permite a la empresa comparar resultados, automatizar recomendaciones y alimentar herramientas de análisis como Power BI sin depender de interpretaciones manuales o texto ambiguo.

#### Definir el esquema de salida (output schema)

````python
# clase base model
class MarketingAnnalysis(BaseModel):
    problema: str = Field(description="principal problema detectado")
    causa: str = Field(description="causa probable del problema")
    accion: str = Field(description="accion recomendada")
    impacto: str = Field(description="nivel de impacto esperado: alto, medio o bajo")
````

#### Crear el parser

````python
parser = JsonOutputParser(pydantic_object=MarketingAnnalysis)
format_instructions = parser.get_format_instructions()
````

#### Crear el prompt

````python
prompt = PromptTemplate(
    template=(
        "Analiza la siguiente situación de marketing.\n"
        "{format_instructions}\n"
        "{input}\n"
    ),
    input_variables=["input"],
    partial_variables={"format_instructions": format_instructions},
)
````

#### Crear el modelo

````python
llm = get_llm(llm="dsk", params={"temperature": 0.5})
````

#### Crear la cadena

````python
chain = prompt | llm | parser
````

#### Ejecutar

````python
response = chain.invoke({"input": "Una tienda online tiene mucho tráfico pero baja conversión."})
print("Resultado de la consulta")
print(f"Problema: {response['problema']}")
print(f"Causa: {response['causa']}")
print(f"Accion: {response['accion']}")
print(f"Impacto: {response['impacto']}")
````

## Document Object y Document Loaders en LangChain

En LangChain, un **Document** es una estructura que representa una unidad de información compuesta por dos elementos: el contenido del texto (`page_content`) y un conjunto de metadatos (`metadata`) que aportan contexto adicional, como el origen, identificadores o fechas. 

Este formato estandarizado permite trabajar de manera uniforme con distintos tipos de datos, ya provengan de archivos, páginas web o bases de datos. Para obtener estos documentos desde fuentes reales, LangChain utiliza los llamados **Document Loaders**, que son componentes diseñados para cargar información desde múltiples orígenes como PDFs, URLs o sitios web y convertirla automáticamente en objetos `Document`. 

De esta forma, los datos externos se integran fácilmente en los flujos de trabajo con modelos de lenguaje, siendo la base para aplicaciones más avanzadas como sistemas de recuperación de información (RAG), análisis de datos o asistentes inteligentes.

### URL and website loader

El **URL and Website Loader** en LangChain permite cargar contenido directamente desde páginas web o URLs y convertirlo automáticamente en objetos `Document`, facilitando así el uso de información online dentro de aplicaciones de IA. 

En lugar de trabajar con archivos locales como PDFs, este tipo de loader accede al contenido web, extrae el texto relevante y lo estructura en el formato estándar de LangChain, incluyendo tanto el contenido (`page_content`) como metadatos asociados, como la fuente o la URL. 

Esto resulta especialmente útil para trabajar con información actualizada, realizar scraping controlado de páginas o integrar datos públicos en sistemas como RAG, chatbots o análisis automatizados, sin necesidad de procesar manualmente el contenido web.

### Divisores de texto (Text splitters)

Después de cargar documentos, se deben transformar para que se adapten mejor a tu aplicación.

Uno de los ejemplos más sencillos de esta transformación es dividir un documento largo en fragmentos más pequeños que puedan ajustarse a la ventana de contexto del modelo. LangChain incluye transformadores de documentos que facilitan tareas como dividir, combinar, filtrar y manipular documentos.

A alto nivel, así es como funcionan los divisores de texto:

* Dividen el texto en pequeños fragmentos con significado semántico (normalmente oraciones).
* Comienzan a combinar estos fragmentos en bloques más grandes hasta alcanzar un tamaño determinado (según una función específica).

Importar librerias

````python
from langchain_core.documents import Document
from langchain_community.document_loaders import PyPDFLoader, WebBaseLoader
from langchain_text_splitters import CharacterTextSplitter, RecursiveCharacterTextSplitter
````

Cargar los documentos y la web

````python
# Cargar el documento (paper) de LangChain
paper_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/96-FDF8f7coh0ooim7NyEQ/langchain-paper.pdf"
pdf_loader = PyPDFLoader(paper_url)
pdf_document = pdf_loader.load()

# Cargar contenido desde la web de LangChain
web_url = "https://python.langchain.com/v0.2/docs/introduction/"
web_loader = WebBaseLoader(web_url)
web_document = web_loader.load()

# Crear dos divisores de texto diferentes
splitter_1 = CharacterTextSplitter(chunk_size=300, chunk_overlap=30, separator="\n")
splitter_2 = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50, separators=["\n\n", "\n", ". ", " ", ""])

# Aplicar ambos divisores al documento PDF
chunks_1 = text_splitter.split_documents(pdf_document)
chunks_2 = text_splitter.split_documents(web_document)

# Definir una función para mostrar estadísticas de los documentos
def display_document_stats(docs, name):
    """Muestra estadísticas sobre una lista de fragmentos de documentos"""
    total_chunks = len(docs)
    total_chars = sum(len(doc.page_content) for doc in docs)
    avg_chunk_size = total_chars / total_chunks if total_chunks > 0 else 0
    
    # Contar las claves únicas de metadata en todos los documentos
    all_metadata_keys = set()
    for doc in docs:
        all_metadata_keys.update(doc.metadata.keys())
    
    # Mostrar estadísticas
    print(f"\n=== Estadísticas de {name} ===")
    print(f"Número total de fragmentos: {total_chunks}")
    print(f"Tamaño medio de fragmento: {avg_chunk_size:.2f} caracteres")
    print(f"Claves de metadata preservadas: {', '.join(all_metadata_keys)}")
    
    if docs:
        print("\nEjemplo de fragmento:")
        example_doc = docs[min(5, total_chunks-1)]  # Obtiene el 5º fragmento o el último si hay menos
        print(f"Contenido (primeros 150 caracteres): {example_doc.page_content[:150]}...")
        print(f"Metadata: {example_doc.metadata}")
        
        # Calcular distribución de tamaños
        lengths = [len(doc.page_content) for doc in docs]
        min_len = min(lengths)
        max_len = max(lengths)
        print(f"Tamaño mínimo de fragmento: {min_len} caracteres")
        print(f"Tamaño máximo de fragmento: {max_len} caracteres")

# Mostrar estadísticas para ambos conjuntos de fragmentos
display_document_stats(chunks_1, "Splitter 1")
display_document_stats(chunks_2, "Splitter 2")
````

## Modelos de embeddings (Embedding models)

Los modelos de embeddings están diseñados para transformar texto en representaciones numéricas (vectores) que capturan su significado semántico. En lugar de trabajar con palabras como texto plano, estos modelos convierten cada fragmento de texto en un punto dentro de un espacio vectorial. 

Esto permite comparar textos no por coincidencia exacta de palabras, sino por similitud de significado, lo que resulta clave en tareas como búsqueda semántica, recomendación de contenido o sistemas RAG. Gracias a los embeddings, es posible identificar qué textos son más parecidos entre sí midiendo la distancia entre sus vectores. 

Existen múltiples proveedores de modelos de embeddings, como IBM, OpenAI o Hugging Face, y en este contexto se utilizará embeddings locales con Ollama en nuestro entorno de practicas hibrido para generar y trabajar con estas representaciones vectoriales.

### **Vector stores (almacenamiento vectorial)**

Una de las formas más comunes de trabajar con datos no estructurados en sistemas de inteligencia artificial es transformar el texto en **embeddings** (vectores numéricos) y almacenarlos para poder consultarlos posteriormente. En este enfoque, cada fragmento de texto se convierte en un vector dentro de un espacio semántico, y cuando el usuario realiza una consulta, esa consulta también se convierte en un vector. A partir de ahí, el sistema puede recuperar los textos más relevantes encontrando aquellos vectores que son más similares al de la consulta.

Aquí es donde entran en juego los **vector stores**. Un vector store es una base de datos especializada que permite almacenar estos embeddings y realizar búsquedas eficientes basadas en similitud vectorial. En lugar de buscar por palabras clave exactas, como en los sistemas tradicionales, se realiza una búsqueda semántica, lo que mejora significativamente la calidad de los resultados en aplicaciones como RAG, chatbots o sistemas de recomendación.

Existen múltiples opciones de vector stores, tanto locales como en la nube. En este contexto se utiliza **Chroma**, una solución ligera, fácil de integrar y especialmente adecuada para entornos locales como el que estás construyendo con Ollama. Chroma permite almacenar embeddings, consultarlos rápidamente y gestionar colecciones de documentos sin necesidad de infraestructuras complejas, lo que lo convierte en una excelente opción para desarrollo y pruebas de sistemas RAG.


Importar librerias 

````python
from langchain_community.vectorstores import Chroma
````

### Retrievers (recuperadores de información)

Una vez que ya tienes un **vector store funcionando**, el siguiente paso es introducir los retrievers, que son el componente encargado de **recuperar la información relevante para una consulta**. En LangChain, un retriever actúa como una capa intermedia entre tu base de datos vectorial y el modelo de lenguaje, estandarizando la forma en la que se obtienen los documentos más relevantes.

En esencia, un retriever no es más que una interfaz que encapsula la lógica de búsqueda. En el caso de los vector store-backed retrievers, estos utilizan directamente el vector store (como Chroma) para realizar búsquedas semánticas. Internamente, emplean métodos como similarity search o MMR (Maximum Marginal Relevance) para encontrar los fragmentos de texto más cercanos a la consulta en el espacio vectorial.

La principal ventaja de usar un retriever es que abstrae la complejidad del vector store y permite integrarlo fácilmente en pipelines más avanzados, como cadenas (chains) o sistemas RAG. Además, ofrece mayor flexibilidad, ya que puedes ajustar parámetros como el número de resultados (k) o el tipo de búsqueda sin modificar la base de datos subyacente.

Usa el vector store docsearch como un retriever. Esto convierte el vector store en una interfaz de recuperación que puede obtener documentos relevantes

````python
retriever = docsearch.as_retriever()
````

Ejecuta el retriever con la consulta "Langchain". Esto hará:
1. Convertir el texto de la consulta "Langchain" en un vector de embedding
2. Realizar una búsqueda de similitud en el vector store usando ese embedding
3. Devolver los documentos más similares semánticamente a la consulta

````python
docs = retriever.invoke("Langchain")
````

Accede al primer documento (el más relevante) de los resultados. Esto devuelve el objeto Document completo, incluyendo:
 - page_content: el contenido de texto del documento
 - metadata: metadatos asociados como fuente, número de página, etc.

#### Parent document retrievers

Al dividir documentos para su recuperación, suelen existir **objetivos en conflicto:**

- Quieres documentos pequeños para que sus embeddings reflejen con mayor precisión su significado. Si los documentos son demasiado largos, los embeddings pueden perder calidad semántica.
- Quieres documentos lo suficientemente largos como para mantener el contexto de cada fragmento de texto.

El `ParentDocumentRetriever` resuelve este equilibrio dividiendo y almacenando fragmentos pequeños de datos. Durante la recuperación, primero obtiene estos fragmentos pequeños, pero luego utiliza sus identificadores para localizar los documentos originales (más grandes) y devuelve esos documentos completos.
